# R - Functions

A function is a set of statements organized together to perform a specific task. R has a large number of in-built functions and the user can create their own functions.

In R, a function is an object so the R interpreter is able to pass control to the function, along with arguments that may be necessary for the function to accomplish the actions.

The function in turn performs its task and returns control to the interpreter as well as any result which may be stored in other objects.

# Function Definition


 An R function is created by using the <b>keyword</b> function. The basic syntax of an R function definition is as follows

#####  function_name <- function(arg_1, arg_2, ...) {
   ##### Function body 
##### }

## Function Components

```r
display_png(file='Images/r-components-of-functions.png',width = 200,height=200)
```

The different parts of a function are:<br>
<b>Function Name</b> − This is the actual name of the function. It is stored in R environment as an object with this name.<br>

<b>Arguments</b> − An argument is a placeholder. When a function is invoked, you pass a value to the argument. Arguments are optional; that is, a function may contain no arguments. Also arguments can have default values.<br>

<b>Function Body</b> − The function body contains a collection of statements that defines what the function does.<br>

<b>Return Value</b> − The return value of a function is the last expression in the function body to be evaluated.<br>

R has many in-built functions which can be directly called in the program without defining them first. We can also create and use our own functions referred as <b>user defined functions</b>.

# Built-in Function

Simple examples of in-built functions are <b>seq(), mean(), max(), sum(x) and paste(...) </b>etc. They are directly called by user written programs. You can refer most widely used R functions

```r

# Create a sequence of numbers from 32 to 44.
print(seq(32,44))
```

```r
# Find mean of numbers from 25 to 82.
print(mean(25:82))
```

```r
# Find sum of numbers frm 41 to 68.
print(sum(41:68))
```

# User-defined Function

We can create user-defined functions in R. They are specific to what a user wants and once created they can be used like the built-in functions

```r
afunction <- function(a){
    for(i in 1:a){
        b <- i^2
        print(b)
    }
}
```

```r
afunction(6)
```

## Calling a Function without an Argument


```r
newfun <-function(){
    for(i in 1:5){
        b <- i^2
        print(b)
    }
}
```

```r
newfun()
```

## Calling a Function with Argument Values (by position and by name)

The arguments to a function call can be supplied in the same sequence as defined in the function or they can be supplied in a different sequence but assigned to the names of the arguments.

```r
newf <-function(a,b,c){
    result <- a * b +c
    print(result)
}
newf(1,2,45)
```

```r
newf(a =100,b =21,c=32)
```

## Calling a Function with Default Argument

We can define the value of the arguments in the function definition and call the function without supplying any argument to get the default result. But we can also call such functions by supplying new values of the argument and get non default result.

```r
defa <- function(a=4,b=9,c=10){
    result <- a* b + c
    print(result)
}
```

```r
defa()
```

```r
defa(12,12,11)
```

### Lazy Evaluation of Function

Arguments to functions are evaluated lazily, which means so they are evaluated only when needed by the function body.

```r
new.function <- function(a, b) {
   print(a^2)
   print(a)
   print(b)
}
```

```r
new.function(5)
```
